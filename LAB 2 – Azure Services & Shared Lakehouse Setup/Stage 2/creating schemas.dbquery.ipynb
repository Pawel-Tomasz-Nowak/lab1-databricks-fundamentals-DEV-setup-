```sql
-- To avoid hardcoding, we'll use dynamic SQL parameters
DECLARE OR REPLACE var_storage_account STRING = TRIM(:storage_account_name);
DECLARE OR REPLACE var_container STRING = TRIM(:container_name);

-- Create Bronze Schema
EXECUTE IMMEDIATE "CREATE SCHEMA IF NOT EXISTS dbr_dev." || var_container || "_bronze MANAGED LOCATION 'abfss://" || var_container || "@" || var_storage_account || ".dfs.core.windows.net/bronze/'";

-- Create Silver Schema
EXECUTE IMMEDIATE "CREATE SCHEMA IF NOT EXISTS dbr_dev." || var_container || "_silver MANAGED LOCATION 'abfss://" || var_container || "@" || var_storage_account || ".dfs.core.windows.net/silver/'";

-- Create Gold Schema
EXECUTE IMMEDIATE "CREATE SCHEMA IF NOT EXISTS dbr_dev." || var_container || "_gold MANAGED LOCATION 'abfss://" || var_container || "@" || var_storage_account || ".dfs.core.windows.net/gold/'";
```
