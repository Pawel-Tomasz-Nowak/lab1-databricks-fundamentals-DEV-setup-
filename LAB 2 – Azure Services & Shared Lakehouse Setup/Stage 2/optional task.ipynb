# The medallion architecture
This is the modern industry-standard design pattern for organizing data within a Lakehouse. It structures files into three distinct logical layers:

- Bronze: Ingests raw data directly from your source systems (like files or APIs) with no transformations. We keep this as our "immutable source of truth".
- Silver: Filters out corrupt files, removes duplicate records, enforces schemas, and normalizes values.
- Gold: Stores high-level, aggregate data optimized for dashboards or reporting.

# Managed Identity vs. Service Principal

When connecting to cloud storage, organizations typically use either a Service Principal or Managed Identity for authentication.

Service Principal:

- Credential Handling: Requires manual setup of Client IDs, Tenant IDs, and Client Secrets.
- Maintenance Effort: Teams must store credentials securely and monitor their expiration to avoid disruptions.
- Governance Granularity: Access permissions are generally set at the resource or container level.

Managed Identity:

- Credential Handling: Azure automatically manages authentication, eliminating the need for manual credentials.
- Maintenance Effort: No passwords or secrets to manage, reducing the risk of credential leaks or expiration.
- Governance Granularity: Permissions can be assigned directly to specific assets, offering more precise control and integration with Unity Catalog.


# External libraries
If you need extra Python tools or drivers for your PySpark jobs, here’s how you can add them:

- Wheels: Build your own Python packages, upload them to a Volume, and install with %pip install /Volumes/....
- PIP (%pip): Use this to grab standard Python packages from PyPI (like scikit-learn) for your notebook.
- Maven: For Scala or Java libraries—such as JDBC drivers for databases like PostgreSQL or MongoDB—attach them at the cluster level.
